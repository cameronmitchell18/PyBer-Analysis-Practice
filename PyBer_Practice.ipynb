```python
import matplotlib

import matplotlib.pyplot as plt 
```

```python
savings= [0, 100, 150, 300, 430, 500, 203, 319, 800, 970, 450, 1000]
months = ['Jan', 'Feb' , 'Mar' , 'Apr' , 'May' , 'Jun' , 'Jul' , 'Aug' , 'Sep' , 'Oct' , 'Nov' , 'Dec']

plt.plot(months,savings)
plt.xlabel('Month') # This takes in a single argument. Should normally be a string. 
plt.ylabel('Savings ($)') # This takes in a single argument. Should normally be a string.
plt.title('Savings Account Results') # This takes in a single argument. Should normally be a string.
plt.show()

# NOTE: The 'y' axis reprsents the savings and matplotlib auto generates the 'x' axis for us. 

# NOTE: The 'x' axis must always come first when passing it through plt.plot 

# NOTE: Plt.show() should always be the last line of code in order for all of the data to be respresented on the chart. 
```
