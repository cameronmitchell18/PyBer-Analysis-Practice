## Create a Line Chart Using the MATLAB Method

```python
%matplotlib inline

import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"] # Months 

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09] # Dollars in total fares 

# Create the plot
plt.plot(x_axis, y_axis)
```

## Create a Line Chart Using the Object-Oriented Interface

```python
# NOTE: When we switch from using MATLAB to the object-oriented interface method, we use ax.plot() instead of plt.plot().
    # To create a figure, use fig, ax = plt.subplots().
    # To define the axes, use ax.plot().
    # Inside the parentheses of ax.plot(), add the axes separated by a comma.

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# NOTE: Let's break down what the code is doing:
    # The plt.subplots() function returns a tuple that contains a figure and axes object(s).
    # fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
    # The first variable, fig, references the figure.
    # The second variable, ax, references the axes of the figure and allows us to annotate the figure.
    # The ax.plot(x_axis, y_axis) plots the data from the x and y axes.

```

```python
# NOTE: To declare the parameters for line color, width, and marker, we use color=, width=, and marker=, respectively, inside the plt.plot() function.
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

## Skill Drill

```python
# Create the plot and add a label for the legend.
# 1. Change the line color to green.
# 2. Change the marker to a diamond.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

## Create a Vertical Bar Chart Using the MATLAB Method.

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.bar(x_axis, y_axis) # plt.bar will create a bar chart. 
```

```python
# NOTE: If you're graphing time-series data or ordinal values (e.g., age ranges, salary ranges, or groups), a vertical bar chart is a good choice.
# NOTE: We can annotate the bar chart like we annotated the line chart. However, we won't need the marker or linewidth attributes.
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

## Create a Vertical Bar Chart

```python
# NOTE: If you're graphing nominal variables (e.g., favorite music groups among teens, number of votes by county or state) where you sort the data from greatest to least or least to greatest, it's best to use a horizontal bar chart.
# Create the plot
plt.barh(x_axis, y_axis) # To get a horizontal bar chart use barh(horizontal)
```

```python
# NOTE: If you want the data on opposite axes, switch the arguments in the barh() function and run the cell again:
# Create the plot
plt.barh(y_axis, x_axis)
```
