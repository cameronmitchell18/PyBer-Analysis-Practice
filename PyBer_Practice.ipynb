## Create a Line Chart Using the MATLAB Method

```python
%matplotlib inline

import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"] # Months 

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09] # Dollars in total fares 

# Create the plot
plt.plot(x_axis, y_axis)
```

## Create a Line Chart Using the Object-Oriented Interface

```python
# NOTE: When we switch from using MATLAB to the object-oriented interface method, we use ax.plot() instead of plt.plot().
    # To create a figure, use fig, ax = plt.subplots().
    # To define the axes, use ax.plot().
    # Inside the parentheses of ax.plot(), add the axes separated by a comma.

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# NOTE: Let's break down what the code is doing:
    # The plt.subplots() function returns a tuple that contains a figure and axes object(s).
    # fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
    # The first variable, fig, references the figure.
    # The second variable, ax, references the axes of the figure and allows us to annotate the figure.
    # The ax.plot(x_axis, y_axis) plots the data from the x and y axes.

```

```python
# NOTE: To declare the parameters for line color, width, and marker, we use color=, width=, and marker=, respectively, inside the plt.plot() function.
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

## Create a Vertical Bar Chart Using the MATLAB Method.

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.bar(x_axis, y_axis) # plt.bar will create a bar chart. 
```

```python
# NOTE: If you're graphing time-series data or ordinal values (e.g., age ranges, salary ranges, or groups), a vertical bar chart is a good choice.
# NOTE: We can annotate the bar chart like we annotated the line chart. However, we won't need the marker or linewidth attributes.
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

## Create a Vertical Bar Chart

```python
# NOTE: If you're graphing nominal variables (e.g., favorite music groups among teens, number of votes by county or state) where you sort the data from greatest to least or least to greatest, it's best to use a horizontal bar chart.
# Create the plot
plt.barh(x_axis, y_axis) # To get a horizontal bar chart use barh(horizontal)
```

```python
# NOTE: If you want the data on opposite axes, switch the arguments in the barh() function and run the cell again:
# Create the plot
plt.barh(y_axis, x_axis)

# NOTE: The horizontal bar chart is not a good way to represent the data because the bar should convey an amount on the x-axisnot a date.
```

```python
# NOTE: To invert the y-axis to have the months in ascending order, use the gca() method. The gca() method means "get current axes." 
# We can chain the gca() method to the invert_yaxis() method by using gca().invert_yaxis(), as shown here:

# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

### Create a Scatter Plot Using the MATLAB Method

There are two ways to create scatter plots using the MATLAB approach:

plt.plot()
plt.scatter()

```python
# NOTE: When we use the plt.plot() function to create a scatter plot, we need to add a lowercase "o" as a parameter inside the parentheses. This switches the plot from a line chart to a scatter plot.
# 
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, 'o')
```

```python
# NOTE: To create a scatter plot using the plt.scatter() function, add the x-axis and y-axis parameters inside the parentheses

plt.scatter(x_axis, y_axis)
```

### Create a Bubble Chart Using the MATLAB Method

Bubble charts are useful when presenting financial, population, and weather data because you can add a third and fourth factor to convey more information.

By changing the "dot" into a "bubble," we are adding a third factor: size.

If there is more than one dataset that uses the same axes, we can change the color of each marker for each dataset, which will add a fourth factor: color.

```python
# NOTE: Next, to change the size of our marker on the graph, we have to use the s, or size parameter, for the marker.
# NOTE: The s parameter must be equal to a feature of our data.
# NOTE: We will make the size of each marker equal to the data floating-point decimal values of the y-axis by using s=y_axis.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# NOTE: We can adjust the size by multiplying the data in the y-axis by any value.

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# NOTE: You can use list comprehension to replace many for and while loops. 
# List comprehension is faster because it is optimized for Python to spot a predictable pattern during looping.

# EXAMPLE: new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```
